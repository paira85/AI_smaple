```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#데이터세팅
height = [155.3,157.5,156.5,163.9,169.3,170.5,195.5,175.1,173.8,177.7,182.6,180.7,186.7,189.9,189.5]
foot_size = [200,230,235,240,245,250,255,260,265,270,275,280,285,290,300]

#이상치 데이터 전처리
height =  height[0:6] + height[7:15]
foot_size = foot_size[0:6] + foot_size[7:15]

#회귀분석과 오차
#오차가 가장 작은 선을 찾는 방법은 "평균제곱오차의 미분 값이 0인 a(기울기)와 b(절편)"을 구하는 것이다.
#공식으로는 최소제곱법이다.

#전체 공식
# a =  (y1 - y(평균)) * ( x1 - x(평균)) /  (x1 - x(평균) 제곱 )
# b =  y(평균) - a(평균)

#최소제곱법 구현
x = np.array(foot_size)
y = np.array(height)
print('####################')
a = np.sum(y - np.mean(y))
print('Y의 평균         = ' , np.mean(y))
print('Y - 첫번재 오차  = ' , np.mean(y) - y[0])
print('키의 전체 오차   = '  , y - np.mean(y))

b = np.sum(y - np.mean(x))
print('x의 평균         = ' , np.mean(x))
print('x - 첫번재 오차  = ' , np.mean(x) - x[0])
print('발   전체 오차   = '  , x - np.mean(x))


print(( y - np.mean(y) ) * ( x - np.mean(x) ))
# aa(기울기) 의 분자
aa = np.sum(( y - np.mean(y) ) * ( x - np.mean(x) ))
print('분자 ' , aa)
print('####################')

# bb(기울기 )의 분모
print('발의제곱 ' , ( x - np.mean(x))**2)
bb = np.sum(( x - np.mean(x))**2)
print('분모 ' , bb)

## cc= 기울기 계산값
cc = aa / bb
## dd = 절편
dd = np.mean(y) - (cc * np.mean(x))
print('기울기 ' , cc  , '절편 ' , dd )


##회귀선 구하기
#최솟값 - 최댓값 사이에 수를 1cm간격으로 저장한다.
line_x = np.arange(min(x) , max(x) , 1)
print('line_x ', line_x)
line_y = cc * line_x + dd

#그래프 결과
plt.plot(line_x , line_y , color = 'r')
plt.scatter(x , y)
plt.xlabel('발크기(foot_size)')
plt.ylabel('키크기(height)')
plt.show()

#########검증

#오차 = 실제값 - 예측값
#오차 = y - (기울기 * x  + 절편)
#오차 = y - (cc(기울기) * x + dd(절편))
#오차의 제곱의 합  np.sum((y - (cc * x + dd))**2)
mse_sum = np.sum((y - (cc * x + dd))**2)
print('오차의 제곱의 합 ' , mse_sum)
print('mse ' , np.mean(y)  - cc * np.mean(x))
```
